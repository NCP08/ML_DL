# 훈련 세트와 테스트 세트
- 훈련한 데이터를 다시 테스트에 사용하는 것은 정확한 학습되었는지 판단하기 어렵다
- 그러므로 훈련 데이터로 학습을 진행하고, 테스트는 다른 데이터로 점수를 내야 한다

```python
fish_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0,
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0,
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0, 9.8,
                10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
fish_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0,
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0,
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0, 6.7,
                7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]
```

```python
print(len(fish_length))
print(len(fish_weight))
```

```python
# 입력 데이터
fish_data = [[l, w] for l, w in zip(fish_length, fish_weight)]

print(fish_data)
print(len(fish_data))
```

```python
# 타겟 데이터(정답데이터)
fish_target = [1] * 35 + [0] * 14
```

```python
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier()
```

```python
# 인덱스 4의 데이터
print(fish_data[4])
```

```python
# 슬라이싱
print(fish_data[0:5])
print(fish_data[:5])
```

```python
print(fish_data[44:])
```

```python
# 훈련 세트와 테스트 세트로 분리
train_input = fish_data[:35]
train_target = fish_target[:35]

test_input = fish_data[35:]
test_target = fish_target[35:]
```

```python
print(train_input)
print(train_target)
```

```python
print(test_input)
print(test_target)
```

```python
# 이제 훈련 데이터로 학습하고
# 테스트 데이터로 점수 확인하자
kn = kn.fit(train_input, train_target)

kn.score(test_input, test_target)
```

## 정확도가 0.0
- 즉, 0점이다
- 이유는 도미로 학습시키고, kn은 도미로만 분류하는 학습이 되어 있다
- 그러므로 빙어로 테스트를 진행해도, 도미로 분류하므로 0점이 되었다
- '샘플링 편향'



```python
import numpy as np

input_arr = np.array(fish_data)
target_arr = np.array(fish_target)
```

```python
print(input_arr.shape)
print(input_arr)
```

```python
# 데이터의 순서를 바꿀 난수 생성
np.random.seed(42)      # 실행하는 모두가 같은 난수를 갖도록 유도하는 상수값, 실제 혼자할 때는 제거해도 됨
index = np.arange(49)
print(index)

np.random.shuffle(index)    # 인덱스 배열을 섞는다
print(index)
```

```python
# 배열 인덱싱(인덱스 1과 3의 데이터만 추출)
print(input_arr[[1, 3]])
```

```python
# 난수로 생성한 인덱스를 주면 해당 데이터가 추출된다
train_input = input_arr[index[:35]]
train_target = target_arr[index[:35]]
```

```python
# 테스트 데이터 준비
test_input = input_arr[index[35:]]
test_target = target_arr[index[35:]]

print(test_target)
```

```python
# 시각화해서 잘 섞인 것을 확인할 수 있다
import matplotlib.pyplot as plt

# plt.rcParams['figure.figsize'] = [10, 7]

plt.scatter(train_input[:, 0], train_input[:, 1], c='blue')
plt.scatter(test_input[:, 0], test_input[:, 1], c='gold')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
# 잘 섞인 훈련데이터로 학습 진행
kn = kn.fit(train_input, train_target)
```
