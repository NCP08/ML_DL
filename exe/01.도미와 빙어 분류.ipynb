# 도미와 빙어의 생선 분류 문제

# 프로그래밍 방식


```python
#도미의 길이와 무게 정보
bream_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0, 31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0, 35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0]
bream_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0, 500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0, 700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0]
```

```python
print(bream_length)
print(bream_weight)
print(len(bream_length))
print(len(bream_weight))
```

```python
import matplotlib.pyplot as plt

plt.scatter(bream_length, bream_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
# 빙어 길이와 무게
smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]
```

```python
print(smelt_length)
print(smelt_weight)
print(len(smelt_length))
print(len(smelt_weight))
```

```python
plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
# 도미와 빙어 합치기
length = bream_length + smelt_length
weight = bream_weight + smelt_weight

print(length)
print(weight)
print(len(length))
print(len(weight))
```

# 리스트 내포, zip함수
https://wikidocs.net/32#zip

```python
# 입력데이터 (길이, 무게), 타겟데이터(도미=1 or 빙어=0)
# 각각 나눠진 2개의 데이터를 묶는 전처리
fish_data = [[l, w] for l, w in zip(length, weight)]

print(fish_data)
```

```python
# 순서대로 합쳤으므로 타겟데이터는 도미가 35개, 빙어가 14개되어 있다
fish_target = [1] * 35 + [0] * 14

print(fish_target)
```

# k 최근접 이웃 알고리즘


```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
# 머신러닝 알고리즘 객체 준비
kn = KNeighborsClassifier()

# 학습(입력데이터, 타겟데이터(정답))
kn.fit(fish_data, fish_target)
```

```python
# 점수(맞춰보자)
kn.score(fish_data, fish_target)
```

# 위의 1.0은 모두 맞췄다(100%)

```python
# 새로운 데이터를 넣었을 때 어떤 것으로 분류할 것인지
# 1이니까 도미로 분류
kn.predict([[30, 600]])
```

```python
plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.scatter(30, 600, marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
print(kn._fit_X)
```

```python
print(kn._y)
```

```python
# 판별할 때 가장 가까운 49개를 참조해서 분류해라
kn49 = KNeighborsClassifier(n_neighbors=49)
```

```python
# 49개의 데이터를 참조하므로, 다수결에 의해 모두 도미로 학습된다.
kn49.fit(fish_data, fish_target)
```

```python
# 그러므로 다시 측정해보면 빙어도 도미로 분류되므로 틀렸기 때문에
# 35/49 = 0.7142857142857143
kn49.score(fish_data, fish_target)
kn49.score(fish_data, fish_target)
```
